```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Categories': ['A', 'B', 'C', 'D', 'E'],
    'Values': [10, 20, 15, 25, 30],
    'Trend': [5, 15, 10, 20, 25]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the figure size
plt.figure(figsize=(12, 6))

# Bar Chart
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.bar(df['Categories'], df['Values'], color='skyblue', label='Values')
plt.title('Bar Chart of Categories')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()

# Line Chart
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.plot(df['Categories'], df['Trend'], marker='o', color='orange', label='Trend')
plt.title('Line Chart of Trend Over Categories')
plt.xlabel('Categories')
plt.ylabel('Trend Values')
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
