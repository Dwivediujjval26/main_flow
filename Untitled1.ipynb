```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data generation
np.random.seed(0)
data = {
    'Variable1': np.random.normal(loc=50, scale=10, size=100),  # Normal distribution
    'Variable2': np.random.normal(loc=30, scale=5, size=100),   # Normal distribution
}
# Introduce some outliers
data['Variable1'][[5, 15, 20]] = [100, 120, 130]  
data['Variable2'][[2, 30, 60]] = [0, 100, -50]  

df = pd.DataFrame(data)

# Set up the figure
plt.figure(figsize=(15, 10))

# Histogram and Box Plot for Variable1
plt.subplot(2, 2, 1)
sns.histplot(df['Variable1'], bins=20, kde=True)
plt.title('Distribution of Variable1')
plt.xlabel('Variable1')

plt.subplot(2, 2, 2)
sns.boxplot(x=df['Variable1'])
plt.title('Box Plot of Variable1')

# Histogram and Box Plot for Variable2
plt.subplot(2, 2, 3)
sns.histplot(df['Variable2'], bins=20, kde=True)
plt.title('Distribution of Variable2')
plt.xlabel('Variable2')

plt.subplot(2, 2, 4)
sns.boxplot(x=df['Variable2'])
plt.title('Box Plot of Variable2')

# Adjust layout
plt.tight_layout()
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap')
plt.show()

# Scatter plot to check for correlation
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Variable1', y='Variable2', data=df)
plt.title('Scatter Plot of Variable1 vs Variable2')
plt.xlabel('Variable1')
plt.ylabel('Variable2')
plt.axhline(0, color='red', linestyle='--', linewidth=1)  # Line for reference
plt.axvline(0, color='red', linestyle='--', linewidth=1)  # Line for reference
plt.show()
```
